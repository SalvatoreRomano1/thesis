## (1) Functions for plots' labels

```python
# Attach a text label above each bar in *rects*, displaying its height.
def autolabel_bar1(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                ha='center', va='bottom')

def autolabel_bar(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

#use this to plot percentage
def autolabel_barPC(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}%'.format(height*1),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
#use this to plot WITH 2 GRAPHS
def autolabel_barPC1(rects):
    for rect in rects:
        height = rect.get_height()
        ax1.annotate('{:.0f}%'.format(height*1),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
def autolabel_barPC2(rects):
    for rect in rects:
        height = rect.get_height()
        ax2.annotate('{:.0f}%'.format(height*1),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

        
        
def autolabel_bar_1(rects):
    for rect in rects:
        height = rect.get_height()
        ax1.annotate('{:.0f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
def autolabel_bar_2(rects):
    for rect in rects:
        height = rect.get_height()
        ax2.annotate('{:.0f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        

#def autolabel_barPC(rects):
    #total = sum(map(lambda x: x.get_height(),rects))
    
#   total=0
#   for rect in rects:
#       total+=rect.get_height()
   
#   for rect in rects:
#       height = rect.get_height()
#       ax.annotate('{:.2f}%'.format(height/total),
#                   xy=(rect.get_x() + rect.get_width() / 2, height),
#                   xytext=(0, 3),  # 3 points vertical offset
#                   textcoords="offset points",
#                   ha='center', va='bottom')


```

## (3) Function to hide cell

```python
# Use hide_toggle(for_next=True)
# OR hide_toggle()
# to hide a cell


from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)
```
